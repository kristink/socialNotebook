* What is your dataset?
    * The datasets used in the project are 5 in total:
    
        <b>acs2017_census_tract_data:</b>
            https://www.kaggle.com/datasets/muonneutrino/us-census-demographic-data
            This includes population, racial/ethincity, income along with other geographical and demographic data.
            This dataset covers the year 2017.
            The data here were collected by the US Census Bureau.
            The tract data is divided by tract id's where a census tract, census area, census district or 
            meshblock is a geographic region defined for the purpose of taking a census.[1] This dataset is 15 MB and includes 74002 lines and 36 columns of data. 
            
        <b>acs2017_county_data:</b>
            https://www.kaggle.com/datasets/muonneutrino/us-census-demographic-data
            This datasets includes the same information that the census tract data includes but grouped by the boroughs instead of having the details of the census tract. This dataset is 622 KB and includes 3221 lines and 36 columns of data
        
        <b>Public_Recycling_Bins:</b>
            https://data.cityofnewyork.us/Environment/Public-Recycling-Bins/sxx4-xhzg 
            Includes the locations of recycling bins in New York City. This dataset is 47 KB and includes 546 lines and 6 columns of data
        
        <b>Recycling_Diversion_and_Capture_Rates:</b>
            https://data.cityofnewyork.us/Environment/Recycling-Diversion-and-Capture-Rates/gaq9-z3hz 
            This dataset is provided by the City of New York. It includes diversion and capture rates.
            Diversion rate total = Total Recycling / Total Waste
            Capture rate paper = Total Recyclable Paper / Max Recyclable Paper in Waste Stream
            Capture rate mpg = Total Recyclable Metal, Glass, Plastic & Beverage Cartons / Max Recyclable Metal, Glass, Plastic & Beverage Cartons in Waste Stream
            Capture rate total =(Total Recycling - Leaves) / (Max Recyclable Paper + Max Recyclable Metal, Glass, Plastic & Cartons) x 100. This dataset is 145 KB and includes 2833 lines and 10 columns of data. The dataset is partly spatial, since two of it's columns represent the latitute and longitute of the recycling bin locations.
              
        <b>borough_boundaries:</b>
            https://data.cityofnewyork.us/City-Government/Borough-Boundaries/tqmj-j8zm
            This dataset is probided by the Department of City Planning in New York and it represents the boundaries of the boroughs in New York City (water areas excluded). 
            The data is geo data represented in a geojson file that includes the coordinates for each borough, shape area, shape length, the borough code, and borough name.
            
            
        census_block_loc:
        
            This was gotten from the FCC Census Block Conversions API 
            https://www.fcc.gov/general/census-block-conversions-api
            It has census block code for a 200 x 200 grid containing New York City and bit of the surrounding area. It contains 37111 total rows, 41 variables (columns), 2.17 MB

The first thing to do was importing all the data and filtering most of the data to the year 2017.
This action was chosen since the census data was from 2017 while the recycling data covered a span from 2016-2019.
We then drop rows with nan values

```python
# Uploading datasets - using only 2017 unless when looking at recycling over all time
blocks = pd.read_csv('data/census_block_loc.csv')
census_unfiltered = pd.read_csv('data/acs2017_census_tract_data.csv', index_col=0)
census = pd.read_csv('data/acs2017_county_data.csv', index_col=0).dropna()
census_tract = pd.read_csv('data/acs2017_census_tract_data.csv')
census = census.loc[census['State'] == 'New York'] # since we are only interested in NY data
recycling = filter_df(pd.read_csv('data/Recycling_Diversion_and_Capture_Rates.csv'), 'Fiscal Year', 2017).dropna()
recycling_all = pd.read_csv('data/Recycling_Diversion_and_Capture_Rates.csv').dropna()
recycling_bins = pd.read_csv('data/Public_Recycling_Bins.csv').dropna()
recycling_bins['Borough'] = recycling_bins['Borough'].str.strip() # strip whitespaces from columns bc one had a leading white space
```

In our recycling dataset, there was a distinction made between Queens East and Queens West. The same for Brooklyn, but it was divided into Brooklyn North and Brooklyn South. To match these datasets with the rest of our datasets that did not have this property, we changed the naming in the dataset so that Brooklyn South and Brooklyn West would simply be named Brooklyn, same for Queens.

```python
#change f.x. Queens County to Queens so every dataset is the same, done for both the recycling data for 2017 and over all time

recycling['Zone'] = recycling.Zone.apply(lambda x: 'Queens' if 'Queens' in x else x)
recycling['Zone'] = recycling.Zone.apply(lambda x: 'Brooklyn' if 'Brooklyn' in x else x)

recycling_all['Zone'] = recycling_all.Zone.apply(lambda x: 'Queens' if 'Queens' in x else x)
recycling_all['Zone'] = recycling_all.Zone.apply(lambda x: 'Brooklyn' if 'Brooklyn' in x else x)
```

```python
recycling['Date'] = recycling['Fiscal Year'].astype(str)  + '-' + recycling['Fiscal Month Number'].astype(str) + '-01'
recycling['Date'] = pd.to_datetime(recycling['Date'], dayfirst=True)

recycling_all['Date'] = recycling_all['Fiscal Year'].astype(str)  + '-' + recycling_all['Fiscal Month Number'].astype(str) + '-01'
recycling_all['Date'] = pd.to_datetime(recycling_all['Date'], dayfirst=True)
```

To make it easier to use the data, a structure is initialized for all the boroughs. The data from all of our different datasets is then used to fill the datastructure for each boroguh. The information about the boroughs are kept in a list of boroughs. Since some of our datasets included counties and other ones boroughs, we initialized a dictionary to map borough to county, making it easier to index into our datasets in an understandable way and minimizing the likelihood of mistakes. 

```python
# Initialize data structures and variables 
borough_names = ['Bronx', 'Brooklyn', 'Manhattan', 'Queens', 'Staten Island']

borough_county_mapping = {'Bronx': 'Bronx County', 
                         'Brooklyn': 'Kings County', 
                         'Manhattan': 'New York County', 
                         'Queens': 'Queens County', 
                         'Staten Island': 'Richmond County'}

borough_dict = {'name': '', 
               'avg_paper_recycle': 0, 
               'avg_plastic_recycle': 0,
                'avg_total_capture': 0, 
                'avg_total_diversion': 0,
                'total_recycling_bins': 0,
                'total_population_count': 0,
                'income_per_capita': 0,
                'n_men': 0, 
                'n_women': 0,
                'main_gender': '',
                'percentage_hispanic': 0, 
                'percentage_white': 0,
                'percentage_black': 0,
                'percentage_native': 0, 
                'percentage_asian': 0,
                'percentage_pacific': 0,
                'main_ethnicity': ''
               }
boroughs = []

genders = ['Men', 'Women']
ethnicities = ['Hispanic', 'White', 'Black', 'Native', 'Asian', 'Pacific']

diversion_total_col_name = 'Diversion Rate-Total (Total Recycling / Total Waste)'
paper_col_name = 'Capture Rate-Paper (Total Paper / Max Paper)'
plastic_col_name = 'Capture Rate-MGP (Total MGP / Max MGP)'
capture_total_col_name = 'Capture Rate-Total ((Total Recycling - Leaves (Recycling)) / (Max Paper + Max MGP))x100'
```

```python
# Fetching data for each borough and creating a dataset with the information
# we are interested in in this project

for name in borough_names:
    b = copy.deepcopy(borough_dict) # make deep copy so we don't modify the data structure in each loop
    b['name'] = name
    
    temp_recycling_df = recycling.loc[recycling['Zone'] == name]
    
    b['avg_paper_recycle'] = temp_recycling_df[paper_col_name].mean()
    b['avg_plastic_recycle'] = temp_recycling_df[plastic_col_name].mean()
    b['avg_total_capture'] = temp_recycling_df[capture_total_col_name].mean()
    b['avg_total_diversion'] = temp_recycling_df[diversion_total_col_name].mean()
    
    temp_recycling_bin_df = recycling_bins.loc[recycling_bins['Borough'] == name] 
    b['total_recycling_bins'] = temp_recycling_df.shape[0]
    
    temp_census_df = census.loc[census['County'] == borough_county_mapping[name]]
    b['total_population_count'] = list(temp_census_df['TotalPop'])[0]
    b['income_per_capita'] = list(temp_census_df['IncomePerCap'])[0]
    b['n_men'] = list(temp_census_df['Men'])[0]
    b['n_women'] = list(temp_census_df['Women'])[0]
    b['main_gender'] = return_majority(temp_census_df, genders)
    b['percentage_hispanic'] = list(temp_census_df['Hispanic'])[0]
    b['percentage_white'] = list(temp_census_df['White'])[0]
    b['percentage_black'] = list(temp_census_df['Black'])[0]
    b['percentage_native'] = list(temp_census_df['Native'])[0]
    b['percentage_asian'] = list(temp_census_df['Asian'])[0]
    b['percentage_pacific'] = list(temp_census_df['Pacific'])[0]
    b['main_ethnicity'] = return_majority(temp_census_df, ethnicities)
    boroughs.append(b)
```

### Hypothesis H1: there is a statistically significant difference in income for men and women 

We came up with the hypothesis and decided to do a t-test to see if our hypothesis is correct or not. 

```python
pd.options.mode.chained_assignment = None  # silencing CopyWarning

blocks = blocks[blocks.County.isin(['Bronx','Kings','New York','Queens','Richmond'])]
blocks['Tract'] = blocks.BlockCode // 10000
blocks = blocks.merge(census_unfiltered, how='left', right_index=True,left_on='Tract')
blocks.Income = pd.to_numeric(blocks.Income, errors='coerce')
```

```python
# Make dataframes where first is counties w. more women and second is counties w. more men
female = blocks.loc[blocks['Men'] < blocks['Women']]
male = blocks.loc[blocks['Men'] > blocks['Women']]

female = female.dropna()
male = male.dropna()

# Calc average income
female.Income.mean(), male.Income.mean()
```

Here, we need to debias by lowering the number of samples of counties with more women by randomly selecting data samples in the range of the number of samples for counties with more men. 

```python
female_resampled_income = np.random.choice(female.Income, male.shape[0])
stats.ttest_ind(female_resampled_income, male.Income.values)
```

Results from the t-test show that the p value is much lower than 0.05 and that proves that H1 is correct and that there is in fact statistical difference between income for men and women. 

Now, let's plot two histograms of the incomes for men and women to understand our data better. To calculate the number of bins, the following method was used: $bins = 1 + ceil(log2(n))$ which is based on: https://medium.datadriveninvestor.com/how-to-decide-on-the-number-of-bins-of-a-histogram-3c36dc5b1cd8 
It shows that the curves are slightly curved, showing that there are slightly more people with lower salaries.

```python
bin_count1 = int(np.ceil(np.log2(len(female_resampled_income))) + 1)
bin_count2 = int(np.ceil(np.log2(len(male.Income.values))) + 1)

bin_count1, bin_count2
```

```python
def print_income_genders():
    fig, axs = plt.subplots(1, 2);
    fig.set_size_inches(18.5, 10.5);
    axs[0].hist(female_resampled_income, density=False, bins=13, color = '#A6CEE3');  # density=False would make counts
    axs[1].hist(male.Income.values, density=False, bins=13, color = '#FDBF6F');  # density=False would make counts
    axs[1].set(xlabel='Median Household Income [$]', ylabel='Female Count');
    axs[0].set(xlabel='Median Household Income [$]', ylabel='Male Count');
    axs[0].xaxis.label.set_size(16)
    axs[0].yaxis.label.set_size(16)
    axs[1].xaxis.label.set_size(16)
    axs[1].yaxis.label.set_size(16)
    axs[0].axis([0, 220000, 0, 1000])
    axs[1].axis([0, 220000, 0, 1000])

    plt.setp(axs[0].get_xticklabels(), fontsize=13)
    plt.setp(axs[0].get_yticklabels(), fontsize=13)
    plt.setp(axs[1].get_xticklabels(), fontsize=13)
    plt.setp(axs[1].get_yticklabels(), fontsize=13)
    plt.suptitle('Median Household Income by female and male data',fontsize=25)
    fig.tight_layout()

    plt.savefig('female_male_income.png', bbox_inches='tight');
```

```python
print_income_genders()
```

## Let's use some machine learning

Here we take the columns that we want to use to predict and split the data up in train and test. We wanted to get a clear understanding of which factors that don't affect income directly (like employment f.x.) could still impact your income indirectly. Our test data is 30% of our total dataset while our train data is 70%. Throughout we use random seed 42. 

```python
temp_census_df = temp_census_df.drop(columns=['Unemployment', 'SelfEmployed', 'Employed', 'TractId', 'State', 'County', 'Income', 'IncomeErr', 'IncomePerCapErr', 'TotalPop', 'VotingAgeCitizen', 'Production', 'Construction', 'Service', 'Office', 'Professional', 'Poverty', 'ChildPoverty', 'MeanCommute', 'Men', 'Women'])

# Saving feature names for later use
y = np.array(temp_census_df['IncomePerCap'])
census_df_all = temp_census_df
temp_census_df = temp_census_df.drop('IncomePerCap', axis = 1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(temp_census_df, y, test_size = 0.3, random_state = 42)
X_train.shape, y_train.shape;
```

Here we instantiate the model and fit the data to it. Then we train the model and print the score. We are using a Random Forest regression model, that fits a number of classifying decision tress on various sub-samples of the dataset. It uses averageing to control over-fitting and to improve it's accuracy. 

```python
# Instantiate model with 1000 decision trees
rf1 = RandomForestRegressor(max_depth=6, random_state = 42)

# Train the model on training data
rf1.fit(X_train, y_train);

# Run the test data through the trained model and display the score
print('Score :', round(rf1.score(X_test, y_test), 3))
```

Here we search for the best training parameters. We are using randomized search from Scikit-Learn, which is unlike grid search in the way that not all parameter values are tried out but a fixed number of parameter settings is sampled from the specified distributions.

```python
#Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1000)]

# Number of features to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(1, 30)] + [None]

# Minimum number of samples required to split a node
min_samples_split = [2, 3, 4, 5, 6, 7]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 3, 4, 5,]

# Calculate the best performing pick of hyperparameters available for the model
distributions = {   'n_estimators': n_estimators,
                    'max_features': max_features,
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'min_samples_leaf': min_samples_leaf
                    }

cv = RandomizedSearchCV(rf1, distributions, random_state=42)
search = cv.fit(X_train, y_train)
search.best_params_
```

We fit a new model with the optimized parameters

```python
# Instantiate model with 1000 decision trees
rf1 = RandomForestRegressor(max_depth=search.best_params_['max_depth'], 
                            random_state = 42,
                            n_estimators = search.best_params_['n_estimators'],
                            min_samples_split = search.best_params_['min_samples_split'],
                            min_samples_leaf = search.best_params_['min_samples_leaf'],
                            max_features = search.best_params_['max_features'])

# Train the model on training data
rf1.fit(X_train, y_train);

# Run the test data through the trained model and display the score
print('Score :', round(rf1.score(X_test, y_test), 3))
```

This is to print the plot that shows the most important features when predicing the income. It's interesting to see that being white is the most important when it comes to predicting income, second and third most important features are how you transport. 

```python
def print_feature_importance():
    data_color_normalized = [x / max(rf1.feature_importances_) for x in rf1.feature_importances_]
    my_cmap = plt.cm.get_cmap('GnBu')
    colors = my_cmap(data_color_normalized)

    plt.barh(temp_census_df.columns, rf1.feature_importances_, color=colors)
    sm = ScalarMappable(cmap=my_cmap, norm=plt.Normalize(0,max(rf1.feature_importances_)))

    sm.set_array([])
    cbar = plt.colorbar(sm)
    cbar.set_label('Importance', rotation=270,labelpad=25)

    plt.savefig('feature_importance.png', bbox_inches='tight');
```
